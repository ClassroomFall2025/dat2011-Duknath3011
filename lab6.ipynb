```python
class chunhat:
    def __init__(self, dai, rong):
        self.__dai = dai
        self.__rong = rong
    def get_dai(self):
        return self.__dai
    def set_dai(self, dai):
        self.__dai = dai
    def get_rong(self):
        return self.__rong
    def set_rong(self, rong):
        self.__rong = rong
    def nhap_thong_tin(self):
        self.__dai = float(input("nhap chieu dai: "))
        self.__rong = float(input("nhap chieu rong: "))
    def xuat_thong_tin(self):
        print("chu vi hinh chu nhat la: ", (self.__dai + self.__rong)*2)
        print("dien tich hinh chu nhat la: ", self.__dai * self.__rong)

# chuong trinh chinh
chieudai = chunhat("", "")
chieudai.nhap_thong_tin()
chieudai.xuat_thong_tin()
    
```

## Lab 6 - Bài 4

```python
class ql_sv: 
    def __init__(self, ma_sv="", ho_ten="", diem_sv=0):
        self.ma_sv = ma_sv
        self.ho_ten = ho_ten
        self.diem_sv = diem_sv
        self.dssv = []

    def xep_loai(self):
        if self.diem_sv < 5:
            return "yeu"
        elif 5 <= self.diem_sv < 6.5:
            return "trung binh"
        elif 6.5 <= self.diem_sv < 7.5:
            return "kha"
        elif 7.5 <= self.diem_sv < 9:
            return "gioi"
        else:
            return "xuat sac"

    def __str__(self):
        return f"Ma SV: {self.ma_sv}, Ho Ten: {self.ho_ten}, Diem SV: {self.diem_sv}, Xep Loai: {self.xep_loai()}"

    def nhap_sv(self):
        n = int(input("Nhap so luong sinh vien: "))
        for i in range(n):
            print(f"\nNhap thong tin sinh vien thu {i+1}:")
            ma_sv = input("Ma SV: ")
            ho_ten = input("Ho Ten: ")
            diem_sv = float(input("Diem SV: "))
            sv = ql_sv(ma_sv, ho_ten, diem_sv)
            self.dssv.append(sv)

    def xuat_sv(self):
        print("\nDanh sach sinh vien:")
        for sv in self.dssv:
            print(sv)

    def sv_gioi(self):
        print("\nDanh sach sinh vien gioi:")
        ds_gioi = [sv for sv in self.dssv if sv.xep_loai() == "gioi"]
        if ds_gioi:
            for sv in ds_gioi:
                print(sv)
        else:
            print("Khong co sinh vien gioi")

    def sap_xep(self):
        self.dssv.sort(key=lambda sv: sv.diem_sv, reverse=True)
        print("\nDanh sach sinh vien sau khi sap xep theo diem giam dan:")
        for sv in self.dssv:
            print(sv)

    def menu():
        qlsv = ql_sv()
        while True:
            print("\n" + "==="*10 + " MENU " + "==="*10)
            print("1. Nhap danh sach sinh vien")
            print("2. Xuat danh sach sinh vien")
            print("3. Hien thi sinh vien gioi")
            print("4. Sap xep sinh vien theo diem giam dan")
            print("5. Thoat")
            choice = input("Chon chuc nang (1-5): ")
            if choice == '1':
                qlsv.nhap_sv()
            elif choice == '2':
                qlsv.xuat_sv()
            elif choice == '3':
                qlsv.sv_gioi()
            elif choice == '4':
                qlsv.sap_xep()
            elif choice == '5':
                print("Thoat chuong trinh")
                break
            else:
                print("Lua chon khong hop le. Vui long chon lai.")

# Gọi menu
ql_sv.menu()
```
