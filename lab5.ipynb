## Lab 5 - Bài 5: bài tập làm thêm

```python
import math
import datetime

class MayTinhBoTui:
    def __init__(self):
        self.lich_su = []

    # 1. Phép tính cơ bản
    def phep_tinh_co_ban(self):
        print("\n--- PHÉP TÍNH CƠ BẢN ---")
        a = float(input("Nhập số thứ nhất: "))
        b = float(input("Nhập số thứ hai: "))
        print("Chọn phép tính: +, -, *, /")
        op = input("Phép tính: ")
        if op == '+':
            kq = a + b
        elif op == '-':
            kq = a - b
        elif op == '*':
            kq = a * b
        elif op == '/':
            if b != 0:
                kq = a / b
            else:
                print("Không thể chia cho 0!")
                return
        else:
            print("Phép tính không hợp lệ!")
            return
        print("Kết quả:", kq)
        self.lich_su.append(f"{a} {op} {b} = {kq}")

    # 2. Lũy thừa
    def luy_thua(self):
        x = float(input("Nhập x: "))
        y = float(input("Nhập y: "))
        kq = x ** y
        print(f"{x}^{y} = {kq}")
        self.lich_su.append(f"{x}^{y} = {kq}")

    # 3. Căn bậc hai
    def can_bac_hai(self):
        x = float(input("Nhập x: "))
        if x >= 0:
            kq = math.sqrt(x)
            print(f"√{x} = {kq}")
            self.lich_su.append(f"√{x} = {kq}")
        else:
            print("Không thể tính căn của số âm!")

    # 4. Hàm lượng giác
    def ham_luong_giac(self):
        goc = math.radians(float(input("Nhập góc (độ): ")))
        sinx = math.sin(goc)
        cosx = math.cos(goc)
        tanx = math.tan(goc)
        print(f"sin = {sinx:.4f}, cos = {cosx:.4f}, tan = {tanx:.4f}")
        self.lich_su.append(f"sin, cos, tan({math.degrees(goc)}°)")

    # 5. Logarit
    def logarit(self):
        print("1. log10(x)   2. ln(x)   3. log cơ số tùy chọn")
        chon = input("Chọn loại log: ")
        x = float(input("Nhập x: "))
        if x <= 0:
            print("Giá trị x phải > 0")
            return
        if chon == '1':
            kq = math.log10(x)
        elif chon == '2':
            kq = math.log(x)
        elif chon == '3':
            co_so = float(input("Nhập cơ số: "))
            kq = math.log(x, co_so)
        else:
            print("Lựa chọn không hợp lệ!")
            return
        print("Kết quả:", kq)
        self.lich_su.append(f"log = {kq}")

    # 6. Giải phương trình bậc nhất
    def giai_pt_bac1(self):
        a = float(input("Nhập a: "))
        b = float(input("Nhập b: "))
        if a == 0:
            print("Phương trình vô nghiệm!" if b != 0 else "Vô số nghiệm!")
        else:
            x = -b / a
            print(f"Nghiệm x = {x}")
            self.lich_su.append(f"{a}x + {b} = 0 → x = {x}")

    # 7. Giải phương trình bậc 2
    def giai_pt_bac2(self):
        a = float(input("Nhập a: "))
        b = float(input("Nhập b: "))
        c = float(input("Nhập c: "))
        if a == 0:
            print("Đây không phải phương trình bậc 2!")
            return
        delta = b**2 - 4*a*c
        if delta < 0:
            print("Phương trình vô nghiệm")
        elif delta == 0:
            x = -b / (2*a)
            print(f"Nghiệm kép x = {x}")
        else:
            x1 = (-b + math.sqrt(delta)) / (2*a)
            x2 = (-b - math.sqrt(delta)) / (2*a)
            print(f"x1 = {x1}, x2 = {x2}")
        self.lich_su.append(f"PT bậc 2: {a}x² + {b}x + {c}")

    # 8. Xem lịch sử
    def xem_lich_su(self):
        print("\n--- LỊCH SỬ TÍNH TOÁN ---")
        if not self.lich_su:
            print("Chưa có phép tính nào.")
        else:
            for i, item in enumerate(self.lich_su, 1):
                print(f"{i}. {item}")

    # 9. Hiện thời gian
    def hien_thi_thoi_gian(self):
        print("Thời gian hiện tại:", datetime.datetime.now().strftime("%H:%M:%S %d/%m/%Y"))

    # 10. Menu chính
    def menu(self):
        while True:
            print("""
===============================
   MÁY TÍNH BỎ TÚI
===============================
1. Phép tính cơ bản
2. Lũy thừa
3. Căn bậc hai
4. Hàm lượng giác
5. Logarit
6. Giải phương trình bậc nhất
7. Giải phương trình bậc 2
8. Xem lịch sử
9. Xem thời gian hiện tại
0. Thoát chương trình
""")
            chon = input(" Chọn chức năng: ")
            if chon == '1': self.phep_tinh_co_ban()
            elif chon == '2': self.luy_thua()
            elif chon == '3': self.can_bac_hai()
            elif chon == '4': self.ham_luong_giac()
            elif chon == '5': self.logarit()
            elif chon == '6': self.giai_pt_bac1()
            elif chon == '7': self.giai_pt_bac2()
            elif chon == '8': self.xem_lich_su()
            elif chon == '9': self.hien_thi_thoi_gian()
            elif chon == '0':
                print("tam biet")
                break
            else:
                print("Lựa chọn không hợp lệ!")

# --- Chạy chương trình ---
mt = MayTinhBoTui()
mt.menu()
```
