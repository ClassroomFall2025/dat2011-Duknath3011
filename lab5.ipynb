## Lab 5 - B√†i 5: b√†i t·∫≠p l√†m th√™m

```python
import math
import datetime

class MayTinhBoTui:
    def __init__(self):
        self.lich_su = []

    # 1. Ph√©p t√≠nh c∆° b·∫£n
    def phep_tinh_co_ban(self):
        print("\n--- PH√âP T√çNH C∆† B·∫¢N ---")
        a = float(input("Nh·∫≠p s·ªë th·ª© nh·∫•t: "))
        b = float(input("Nh·∫≠p s·ªë th·ª© hai: "))
        print("Ch·ªçn ph√©p t√≠nh: +, -, *, /")
        op = input("Ph√©p t√≠nh: ")
        if op == '+':
            kq = a + b
        elif op == '-':
            kq = a - b
        elif op == '*':
            kq = a * b
        elif op == '/':
            if b != 0:
                kq = a / b
            else:
                print("Kh√¥ng th·ªÉ chia cho 0!")
                return
        else:
            print("Ph√©p t√≠nh kh√¥ng h·ª£p l·ªá!")
            return
        print("K·∫øt qu·∫£:", kq)
        self.lich_su.append(f"{a} {op} {b} = {kq}")

    # 2. L≈©y th·ª´a
    def luy_thua(self):
        x = float(input("Nh·∫≠p x: "))
        y = float(input("Nh·∫≠p y: "))
        kq = x ** y
        print(f"{x}^{y} = {kq}")
        self.lich_su.append(f"{x}^{y} = {kq}")

    # 3. CƒÉn b·∫≠c hai
    def can_bac_hai(self):
        x = float(input("Nh·∫≠p x: "))
        if x >= 0:
            kq = math.sqrt(x)
            print(f"‚àö{x} = {kq}")
            self.lich_su.append(f"‚àö{x} = {kq}")
        else:
            print("Kh√¥ng th·ªÉ t√≠nh cƒÉn c·ªßa s·ªë √¢m!")

    # 4. H√†m l∆∞·ª£ng gi√°c
    def ham_luong_giac(self):
        goc = math.radians(float(input("Nh·∫≠p g√≥c (ƒë·ªô): ")))
        sinx = math.sin(goc)
        cosx = math.cos(goc)
        tanx = math.tan(goc)
        print(f"sin = {sinx:.4f}, cos = {cosx:.4f}, tan = {tanx:.4f}")
        self.lich_su.append(f"sin, cos, tan({math.degrees(goc)}¬∞)")

    # 5. Logarit
    def logarit(self):
        print("1. log10(x)   2. ln(x)   3. log c∆° s·ªë t√πy ch·ªçn")
        chon = input("Ch·ªçn lo·∫°i log: ")
        x = float(input("Nh·∫≠p x: "))
        if x <= 0:
            print("Gi√° tr·ªã x ph·∫£i > 0")
            return
        if chon == '1':
            kq = math.log10(x)
        elif chon == '2':
            kq = math.log(x)
        elif chon == '3':
            co_so = float(input("Nh·∫≠p c∆° s·ªë: "))
            kq = math.log(x, co_so)
        else:
            print("L·ª±a ch·ªçn kh√¥ng h·ª£p l·ªá!")
            return
        print("K·∫øt qu·∫£:", kq)
        self.lich_su.append(f"log = {kq}")

    # 6. Gi·∫£i ph∆∞∆°ng tr√¨nh b·∫≠c nh·∫•t
    def giai_pt_bac1(self):
        a = float(input("Nh·∫≠p a: "))
        b = float(input("Nh·∫≠p b: "))
        if a == 0:
            print("Ph∆∞∆°ng tr√¨nh v√¥ nghi·ªám!" if b != 0 else "V√¥ s·ªë nghi·ªám!")
        else:
            x = -b / a
            print(f"Nghi·ªám x = {x}")
            self.lich_su.append(f"{a}x + {b} = 0 ‚Üí x = {x}")

    # 7. Gi·∫£i ph∆∞∆°ng tr√¨nh b·∫≠c 2
    def giai_pt_bac2(self):
        a = float(input("Nh·∫≠p a: "))
        b = float(input("Nh·∫≠p b: "))
        c = float(input("Nh·∫≠p c: "))
        if a == 0:
            print("ƒê√¢y kh√¥ng ph·∫£i ph∆∞∆°ng tr√¨nh b·∫≠c 2!")
            return
        delta = b**2 - 4*a*c
        if delta < 0:
            print("Ph∆∞∆°ng tr√¨nh v√¥ nghi·ªám")
        elif delta == 0:
            x = -b / (2*a)
            print(f"Nghi·ªám k√©p x = {x}")
        else:
            x1 = (-b + math.sqrt(delta)) / (2*a)
            x2 = (-b - math.sqrt(delta)) / (2*a)
            print(f"x1 = {x1}, x2 = {x2}")
        self.lich_su.append(f"PT b·∫≠c 2: {a}x¬≤ + {b}x + {c}")

    # 8. Xem l·ªãch s·ª≠
    def xem_lich_su(self):
        print("\n--- L·ªäCH S·ª¨ T√çNH TO√ÅN ---")
        if not self.lich_su:
            print("Ch∆∞a c√≥ ph√©p t√≠nh n√†o.")
        else:
            for i, item in enumerate(self.lich_su, 1):
                print(f"{i}. {item}")

    # 9. Hi·ªán th·ªùi gian
    def hien_thi_thoi_gian(self):
        print("Th·ªùi gian hi·ªán t·∫°i:", datetime.datetime.now().strftime("%H:%M:%S %d/%m/%Y"))

    # 10. Menu ch√≠nh
    def menu(self):
        while True:
            print("""
===============================
   üßÆ M√ÅY T√çNH B·ªé T√öI C√Å NH√ÇN
===============================
1. Ph√©p t√≠nh c∆° b·∫£n
2. L≈©y th·ª´a
3. CƒÉn b·∫≠c hai
4. H√†m l∆∞·ª£ng gi√°c
5. Logarit
6. Gi·∫£i ph∆∞∆°ng tr√¨nh b·∫≠c nh·∫•t
7. Gi·∫£i ph∆∞∆°ng tr√¨nh b·∫≠c 2
8. Xem l·ªãch s·ª≠
9. Xem th·ªùi gian hi·ªán t·∫°i
0. Tho√°t ch∆∞∆°ng tr√¨nh
""")
            chon = input("üëâ Ch·ªçn ch·ª©c nƒÉng: ")
            if chon == '1': self.phep_tinh_co_ban()
            elif chon == '2': self.luy_thua()
            elif chon == '3': self.can_bac_hai()
            elif chon == '4': self.ham_luong_giac()
            elif chon == '5': self.logarit()
            elif chon == '6': self.giai_pt_bac1()
            elif chon == '7': self.giai_pt_bac2()
            elif chon == '8': self.xem_lich_su()
            elif chon == '9': self.hien_thi_thoi_gian()
            elif chon == '0':
                print("T·∫°m bi·ªát üëã")
                break
            else:
                print("L·ª±a ch·ªçn kh√¥ng h·ª£p l·ªá!")

# --- Ch·∫°y ch∆∞∆°ng tr√¨nh ---
mt = MayTinhBoTui()
mt.menu()
```
